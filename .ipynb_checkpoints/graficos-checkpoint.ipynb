```python
import pandas as pd
import plotly.express as px
%matplotlib inline
dfm = pd.read_csv('Pesquisa.csv')
```

```python
dfm = dfm.applymap(lambda s:s.lower() if type(s) == str else s)
```

```python
#faixa de conhecimento 1
dfm['pergunta3'] = dfm['pergunta3'].replace(0, '0-3')
dfm['pergunta3'] = dfm['pergunta3'].replace(1, '0-3')
dfm['pergunta3'] = dfm['pergunta3'].replace(2, '0-3')
dfm['pergunta3'] = dfm['pergunta3'].replace(3, '0-3')
#faixa de conhecimento 2
dfm['pergunta3'] = dfm['pergunta3'].replace(4, '4-7')
dfm['pergunta3'] = dfm['pergunta3'].replace(5, '4-7')
dfm['pergunta3'] = dfm['pergunta3'].replace(6, '4-7')
dfm['pergunta3'] = dfm['pergunta3'].replace(7, '4-7')
#faixa de conhecimento 3
dfm['pergunta3'] = dfm['pergunta3'].replace(8, '8-10')
dfm['pergunta3'] = dfm['pergunta3'].replace(9, '8-10')
dfm['pergunta3'] = dfm['pergunta3'].replace(10, '8-10')
```

```python
dfm1 = dfm[dfm['pergunta3'] == '0-3']
def busca_termos(lista,texto):
    cont = 0;
    contacum = 0;
    print("--Contagem dos Termos " + texto);
    for x in lista:
      cont = dfm1['street1'].str.count(x).sum() + dfm1['street2'].str.count(x).sum() + dfm1['street3'].str.count(x).sum() + dfm1['street4'].str.count(x).sum() + dfm1['street5'].str.count(x).sum() + dfm1['street6'].str.count(x).sum() + dfm1['street7'].str.count(x).sum() + dfm1['street8'].str.count(x).sum() + dfm1['street9'].str.count(x).sum() + dfm1['street10'].str.count(x).sum();
      contacum += cont;
      if (cont > 0):
        print(x,cont);
    print('Quantidade de Termos ' + texto + ' encontrado: ',contacum);

#BUSCA POR TERMOS VAGOS

searchVagos = ['perto','longe','grande','pequeno','alto','baixo']
textoVagos = "Vagos"
busca_termos(searchVagos,textoVagos);
    
#BUSCA POR TERMOS QUE DESCREVEM POSICIONAMENTO 

searchPos = ['próximo', 'proximo', 'frente', 'frenta', 'entrada', 'perto', 'pero', 'lado', 'no redor', 'na rua', 'perto', 'esquina', 'as margens', 'antes', 'colado', 'entre' , 'por trás', 'lateral', 'atrás', 'atras', 'nas proximidades', 'vizinho', 'logo após', 'logo apos', 'depois', 'lado', 'entrada']
textoPos = "que descrevem posicionamentos"
busca_termos(searchPos,textoPos);

#BUSCA POR TERMOS QUE DESCREVEM ROTAS

searchRota = ['pega a esquerda', 'pega a direita', 'descendo','sentido', 'após o sinal', 'apos o sinal', 'passando', 'em direcao', 'em direção', 'seguindo reto', 'indo', 'vindo', 'passando pela', 'rua que vai', 'passa pela', 'na subida', 'bifurcação', 'sobe', 'subindo', 'descendo', 'vira a esquerda', 'vira a direita', 'passar', 'na subida', 'a caminho', 'segue reto', 'desce', 'vindo', 'descendo', 'dobra', 'seguir', 'entra na', 'entra no', 'entra em', 'no sinal', 'cruzamento', 'no caminho', 'direcao', 'direção']
textoRota = "que descrevem rotas"
busca_termos(searchRota,textoRota);

#BUSCA POR TERMOS ALTERNATIVOS - VERNACULARES

searchVernacular = ['praça do assalto', 'morgação', 'rua perigosa', 'praça dos hippies', 'praça dos hippie', 'praça dos rips', 'praça q tem os hippies', 'praça da ternura']
textoVernacular = "Vernaculares"
busca_termos(searchVernacular,textoVernacular);

#BUSCA POR TERMOS HISTÓRICOS

searchHistoricos = ['ouro branco','antigo','onigrat','onígrat']
textoHistoricos = "Históricos"
busca_termos(searchHistoricos,textoHistoricos)
```
